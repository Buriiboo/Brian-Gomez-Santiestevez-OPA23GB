```python
# evolve the poke-data into a list of tuples

def read_data_from_file(pika_data_path):
    data_pikachu = []  # List to hold tuples
    with open(pika_data_path, "r") as file:
        for line in file:
            cleaned_line = line.strip().strip("()")  # removing the () from the line
            parts = cleaned_line.split(",")  # splitting the line into parts using comma
            if len(parts) == 2:  # Ensuring the line has exactly 2 parts
                try:
                    width, height = parts
                    width = round(float(width))
                    height = round(float(height))
                    data_pikachu.append((width, height, "Pikachu"))  # appending a tuple
                except ValueError:
                    # In case conversion to float fails
                    print(f"Warning: Could not process line: {line}")
    return data_pikachu


file_path = r"C:\Users\Burai\Downloads\pikachu.txt"
converted_data = read_data_from_file(file_path)

# Print or process your converted data
#print(converted_data)

#Help from the GPT VVV
# Convert each tuple in converted_data to a string and join with new lines
converted_data_str = "\n".join([str(entry) for entry in converted_data])

#print(converted_data_str)
```

After some reviewing i found out i need to remake this data to make it fit into a Linear Discriminant Analysis. 

step 1) find the means of each class (in this case: Pikachu, pichu) 
step 2) maximize the distance between means of two classes 
step 3) minimize the variation within each class



(Source: Mr Mahesh Huddar : https://www.youtube.com/watch?v=wvz_aJA_mfs&t=239s)


```python
import numpy as np

# same as before but for changing the data strukture to width, height, label where label is 0 for pichu and 1 for pikachu

def read_data_from_file(pika_data_path):
    data_pikachu = []  # List to hold tuples
    with open(pika_data_path, "r") as file:
        next(file)  # Skip the first line
        for line in file:
            cleaned_line = line.strip().strip("()")  # removing the () from the line
            parts = cleaned_line.split(",")  # splitting the line into parts using comma
            if len(parts) == 2:  # Ensuring the line has exactly 2 parts
                try:
                    width, height = parts
                    width = round(float(width))
                    height = round(float(height))
                    data_pikachu.append((width, height, 1))  # appending a tuple
                except ValueError:
                    # In case conversion to float fails
                    print(f"Warning: Could not process line: {line}")
    return data_pikachu


file_path = r"C:\Users\Burai\Downloads\pikachu.txt"
converted_data = read_data_from_file(file_path)

# Print or process your converted data
#print(converted_data)

#Help from the GPT VVV
# Convert each tuple in converted_data to a string and join with new lines
converted_data_str = "\n".join([str(entry) for entry in converted_data])

print(converted_data_str)
```
